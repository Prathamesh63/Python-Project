```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Meta Dataset.csv')

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set 'Date' as the index
data.set_index('Date', inplace=True)

# 1. Line Chart of Closing Prices Over Time
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price')
plt.title('Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# 2. Volume Traded Over Time
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Volume'], label='Volume Traded', color='orange')
plt.title('Volume Traded Over Time')
plt.xlabel('Date')
plt.ylabel('Volume Traded')
plt.legend()
plt.grid(True)
plt.show()

# 3. Histogram of Daily Closing Prices
plt.figure(figsize=(14, 7))
plt.hist(data['Close'], bins=50, color='purple')
plt.title('Histogram of Daily Closing Prices')
plt.xlabel('Closing Price (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 4. Moving Average of Closing Prices
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Close Price')
plt.plot(data.index, data['MA50'], label='50-Day Moving Average')
plt.plot(data.index, data['MA200'], label='200-Day Moving Average')
plt.title('Moving Average of Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()

# 5. Box Plot of Prices for Each Year
data['Year'] = data.index.year
plt.figure(figsize=(14, 7))
sns.boxplot(x='Year', y='Close', data=data)
plt.title('Box Plot of Closing Prices for Each Year')
plt.xlabel('Year')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()

# 6. Scatter Plot of Closing Price vs Volume
plt.figure(figsize=(14, 7))
plt.scatter(data['Volume'], data['Close'], alpha=0.5)
plt.title('Closing Price vs Volume')
plt.xlabel('Volume')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```
